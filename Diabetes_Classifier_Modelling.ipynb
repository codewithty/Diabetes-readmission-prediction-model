```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('/content/diabetic_data.csv')
df.head()
```

```python
#replacing ? with null values
df.replace("?", np.nan, inplace=True)
```

```python
df.to_csv('diabetes_cleaned.csv', index=False)
```

```python
df.head()
```

```python
print(df.describe(include='all'))
```

```python
# dropping unneccasry columns
df.drop(['encounter_id', 'patient_nbr', 'weight', 'payer_code', 'medical_specialty'], axis=1, inplace=True)
df.head()
```

```python
df.columns
```

```python
# Flag Patients with Diabetes Diagnosis using  International Classification of Diseases, 9th Revision, Clinical Modification - ICD-9 Codes (250.xx)
def has_diabetes(row):
    return any(str(code).startswith('250') for code in [row['diag_1'], row['diag_2'], row['diag_3']])

df['diabetes_flag'] = df.apply(has_diabetes, axis=1).astype(int)
```

```python
df.head()
```

```python
# Simplify and Encode Categorical Data
categorical_cols = df.select_dtypes(include='object').columns.tolist()
```

```python
categorical_cols
```

```python
#A different encoding method would be used for readmitted
categorical_cols.remove('readmitted')
```

```python
#Importing sklearn libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
#Using label encoding to encode the selected categorical columns
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df.head()
```

```python
# Encode target columns
df['readmitted'] = df['readmitted'].replace({'NO': 0, '>30': 0, '<30': 1})  # binary flag
target = 'diabetes_flag'
```

```python
# Preparing Features
X = df.drop([target], axis=1)
# Drop diagnosis codes from features to avoid data leakage
X = df.drop([target, 'diag_1', 'diag_2', 'diag_3'], axis=1)
y = df[target]
```

```python
# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
# Modelling and predicting
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Check if diagnosis columns are still in features
diagnosis_cols = ['diag_1', 'diag_2', 'diag_3']
print("Diagnosis columns in X?", any(col in X.columns for col in diagnosis_cols))
print("Columns in X that may be leaking the label:", [col for col in X.columns if col in diagnosis_cols])
```

```python
# Saving Results for Power BI
df_pred = X_test.copy()
df_pred['Actual'] = y_test.values
df_pred['Predicted'] = y_pred
df_pred.to_csv("hospital_diabetes_predictions.csv", index=False)
```

```python
# Saving model for reuse
import joblib
joblib.dump(model, "rf_diabetes_hospital_model.pkl")
```
